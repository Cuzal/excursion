# Procesamiento numérico básico: NumPy

NumPy (*Numerical Python*) es considerada la librería base para el trabajo numérico que hagamos con Python, y aunque su nombre es corto, es costumbre utilizar la abreviatura *np* al importarla.

```python
import numpy as np
```

NumPy es una librería para trabajar con *arrays* multidimensionales, algo que seguro conocemos si hemos trasteado con listas anteriormente. Lógicamente, usar arrays sobre las listas clásicas de Python tiene sus ventajas (si no, no habría una librería entera dedicada a ellos, ¿verdad?).

Podemos resumir las ventajas de los arrays en una sola palabra: **rapidez**. Esto se debe a la manera en la que nuestro ordenador almacena la información. Para explicarlo de forma sencilla, en un array, cada campo es representado por un número menor de bytes que el necesario en una lista y, además, se guarda en memoria en celdas contiguas, algo que por lo general no ocurre al manipular listas.

Veamos con un ejemplo que el tiempo requerido para generar una matriz de ceros relativamente grande es mucho menor si usamos arrays:

## Conjunto de Mandelbrot

El conjunto de Mandelbrot es el fractal más estudiado. Los fractales se caracterizan por la invariabilidad total o parcial que presentan en ciertas de sus características con relación a diversas escalas. En otras palabras, al ampliar una zona de la imagen de uno, vuelve a aparecer la imagen original. Y así indefinidamente.

El código proviene de las *Scipy lecture notes 2020*.

```python
from numpy import newaxis

def compute_mandelbrot(N_max, some_threshold, nx, ny):
    # Dividimos el espacio complejo
    x = np.linspace(-2, 1, nx)
    y = np.linspace(-1.5, 1.5, ny)
    
    c = x[:,newaxis] + 1j*y[newaxis,:]
    
    # Iteración de Mandelbrot
    z = c
    for j in range(N_max):
        z = z**2 + c
        
    mandelbrot_set = (abs(z) < some_threshold)
    
    return mandelbrot_set

mandelbrot_set = compute_mandelbrot(50, 50., 601, 401)

plt.figure(figsize=(15,15))
plt.imshow(mandelbrot_set.T, extent=[-2, 1, -1.5, 1.5])
plt.gray()
plt.show()
```

## Integración

El método más utilizado para la integración es *quad()*, dentro del módulo *integrate*.
Veremos como integrar con esta herramienta mediante los siguientes ejemplos. Lo primero será importar el módulo *integrate*.

```python
import scipy.integrate as integrate
```

<ul>
    <li> <b>Integral definida:</b>
</ul
    
\begin{equation}
 f(x) = \int_{x=0}^{x=5} x^2 \cdot dx
\end{equation}

```python
resultado = integrate.quad(lambda x: x ** 2, 0, 5)
print(resultado)
```

Como podemos observar, nos devuelve dos valores:
Por un lado el resultado (izquierda) y por el otro el posible error cometido al realizar la integral (derecha).

Si queremos visualizar únicamente el resultado, bastará con:

```python
print(resultado[0])
```

<ul>
   <li> <b>Integral múltiple:</b>
</ul>  

\begin{equation}
 Área = \int_{y = 0}^{y=1/2}\int_{x=0}^{x=1-2x} xy \cdot dx dy
\end{equation}

En este ejercicio queremos calcular el área mediante una integral doble, resolviéndola con *nquad*.

```python
def f(x, y):
    return x * y

def lim_y():
    return [0, 0.5]

def lim_x(y):
     return [0, 1 - 2 * y]

area = integrate.nquad(f, [lim_x, lim_y])
print (area[0])
```

## Optimización

Un problema de optimización es aquel que trata de seleccionar el mejor elemento de un conjunto con respecto a algún criterio.

Vamos a resolver este ejercicio:

$$min\;\;x_1x_4(x_1 + x_2 + x_3)+ x_3\;\rightarrow\mathrm{Función\; objetivo}\\
\mathrm{s.t}\; x_1 x_2 x_3 x_4 \geq25\;\rightarrow\mathrm{Restricción\; de\; desigualdad}\\
x_1^2+x_2^2+x_3^2+x_4^2\;=\;40\;\rightarrow\mathrm{Restricción\; de\;igualdad}\\
1\leq x_1 x_2 x_3 x_4\leq 5\;\rightarrow\mathrm{Límites}\\
x_0\;=\;(1,5,5,1)\;\rightarrow\mathrm{Estimación\;inicial}$$

El módulo <i>scipy.optimize</i> proporciona algoritmos para la minimización de funciones (escalares o multidimensionales), ajuste de curvas y búsqueda de raíces.

```python
from scipy.optimize import minimize
```

Lo primero que haremos será definir todo lo necesario: el objetivo y las restricciones.

```python
def objetivo(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3] 
    return x1 * x4 * (x1 + x2 + x3) + x3

def restriccion_desigualdad(x):
    return x[0] * x[1] * x[2] * x[3] - 25.0

def restriccion_igualdad(x):
    sum_cuadrados = 40
    for i in range(4):
        sum_cuadrados = sum_cuadrados - x[i]**2
    return sum_cuadrados
```

A continuación definimos un array con las estimaciones iniciales, ahora tieniendo objetivo como una función de x0, intentará darnos un minimo de optimización.

```python
x0 = [1, 5, 5, 1]
print(objetivo(x0))
```

También determinamos los límites que nos proporciona el enunciado. En este caso, 1 y 5.

Después, hacemos l el limite para las cuatro variables que tenemos. Y, finalmente, las restricciones: primero declararemos el tipo al que pertenece en cada caso (igualdad o desigualdad) y la función (anteriormente ya definida, por tanto solo la invocamos).

```python
l = (1.0, 5.0)
lim = (l, l, l, l)
r1 = {'type': 'ineq', 'fun': restriccion_desigualdad}
r2 = {'type': 'eq', 'fun': restriccion_igualdad}
restricciones = [r1,r2]
```

Por último, aplaicamos *minimize*. Debemos introducir la función, la estimación inicial y en este caso por el uso del método de SLSQP (Sequential Least SQuares Programming), debemos introducir los limites y los requistos necesarios.

```python
solucion = minimize(objetivo,x0,method = 'SLSQP',bounds=lim, constraints=restricciones)
print(solucion)
```

A parte del resultado de la optimzación (x), tenemos muchas más varables. Todas ellas son llamables.
Ojo, *Fun* proporciona el valor de la función objetivo.

```python
print(solucion.x)
```

## Interpolación

Una interpolación sirve para extraer puntos nuevos a partir de un conjunto discreto de puntos ya conocidos.

Vamos a realizar un ejemplo de una interpolación sencilla, tanto lineal como cúbica.

```python
from scipy.interpolate import interp1d

np.random.seed(0)

# Generamos los valores de los que disponemos previamente
tiempo = np.linspace(0, 1, 10)
ruido = 1e-1 * (np.random.random(10)*2 - 1)
valores_observados = np.sin(2 * np.pi * tiempo) + ruido

# Extraemos los valores esperados en tiempos distintos a los de las mediciones
nuevos_tiempos = np.linspace(0, 1, 50)

interpolacion_lineal = interp1d(tiempo, valores_observados)
resultados_lineal = interpolacion_lineal(nuevos_tiempos)

interpolacion_cubica = interp1d(tiempo, valores_observados, kind='cubic')
resultados_cubica = interpolacion_cubica(nuevos_tiempos)

plt.figure(figsize=(12, 8))
plt.plot(tiempo, valores_observados, 'o', ms=6, label='Valores observados')
plt.plot(nuevos_tiempos, resultados_lineal, label='Interpolación lineal')
plt.plot(nuevos_tiempos, resultados_cubica, label='Interpolación cúbica')
plt.legend()
plt.show()
```

## Álgebra lineal
Podemos calcular el determinante y la inversa de matrices (arrays) no singulares.

```python
from scipy import linalg

matriz = np.array([[2, 0, 1], [3, 0, 0], [5, 1, 1]])
print(matriz)

linalg.det(matriz)
```

```python
inversa = linalg.inv(matriz)
print(inversa)
```

Podemos comprobar si el producto matricial de la matriz por su inversa es efectivamente la matriz inicial (con una cierta tolerancia).

```python
identidad = np.eye(3)

np.allclose(np.dot(matriz, inversa), identidad)
```

También podemos efectuar operaciones más complicadas. Veamos la descomposición en valores singulares $A = U \Sigma V^T$.

```python
matriz = np.arange(9).reshape((3, 3)) + np.diag([1, 0, 1])
print(matriz)

U, s, V = linalg.svd(matriz)

S = np.diag(s)
SVD = U.dot(S).dot(V)

print('\n',SVD)

np.allclose(SVD, matriz)
```

Y por último veremos la factorización $A = Q R$.

```python
Q, R = linalg.qr(matriz)

QR = Q.dot(R)
print(QR)

np.allclose(Q.dot(R), matriz)
```

## Transformada de Fourier

> *El estudio profundo de la naturaleza es la fuente más fértil de descubrimientos matemáticos.*
                                                                                        - Joseph Fourier (1768-1830)

Una transformada de Fourier es la función matemática empleada para transformar señales entre el dominio del tiempo o el espacio y el dominio de la frecuencia. Además, es reversible, siendo capaz de transformarse en cualquiera de los dominios al otro.

Tiene muchas aplicaciones en la física y la ingeniería, y en lo que queda de 'excursión' va a estar muy presente.

Vamos a realizar un ejemplo completo del trabajo que se suele hacer con una transformada de Fourier en el mundo de las ondas. El objetivo es reconocer las diferentes funciones que generan los sonidos que, combinados, forman la onda sonora que llega hasta nuestros oídos.

Lo primero será generar una señal. Vamos a escoger una señal que provenga de dos ondas relativamente sencillas que además presenta unas perturbaciones aleatorias.

(Adaptado de *Scipy lecture notes 2020*)

```python
from scipy import fftpack

np.random.seed(1234)

time_step = 0.02
period = 5.
time_vec = np.arange(0, 20, time_step)
sig = (np.sin(2 * np.pi / period * time_vec) + np.cos(np.pi/5 * time_vec) + 0.5 * np.random.randn(time_vec.size))

plt.figure(figsize=(12, 6))
plt.xlabel('Tiempo')
plt.ylabel('Presión')
plt.plot(time_vec, sig, label='Señal original')
plt.show()
```

Una vez que tenemos nuestra señal compuesta y enmascarada, vamos a obtener su transformada de Fourier para averiguar de qué frecuencias proviene. En este caso, la gráfica presentará dos picos, pues proviene de un seno y un coseno sumados. Además, los picos de Fourier se dan justo en la frecuencia que presentaba la onda original.

```python
# La transformada de Fourier de la señal
sig_fft = fftpack.fft(sig)

# En valor absoluto ya que la FFT pertenece al plano complejo
power = np.abs(sig_fft)

# Frecuencias correspondientes
sample_freq = fftpack.fftfreq(sig.size, d=time_step)

plt.figure(figsize=(12, 10))
plt.plot(sample_freq, power)
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Coeficiente de Fourier')

# Buscamos picos por encima de 0
pos_mask = np.where(sample_freq > 0)
freqs = sample_freq[pos_mask]
peak_freq = freqs[power[pos_mask].argmax()]

# Comprobamos si coinciden con las frecuencias con las que creamos la señal
print(np.allclose(peak_freq, 1./period))

# Vemos en un recuadro la parte que nos interesa
axes = plt.axes([0.55, 0.3, 0.3, 0.5])
plt.title('Picos de frecuencia')
plt.plot(freqs[:8], power[:8])
plt.setp(axes, yticks=[])
```

Con toda esta información, sabemos que la señal proviene de dos ondas de frecuencias conocidas, por lo que podemos eliminar todas esas perturbaciones y mostrar la señal 'limpia'.

```python
high_freq_fft = sig_fft.copy()
high_freq_fft[np.abs(sample_freq) > peak_freq] = 0
filtered_sig = fftpack.ifft(high_freq_fft)

plt.figure(figsize=(12, 6))
plt.plot(time_vec, sig, label='Señal original')
plt.plot(time_vec, filtered_sig, linewidth=3, label='Señal corregida')
plt.xlabel('Tiempo [s]')
plt.ylabel('Presión')
plt.legend(loc='best')
plt.show()
```

### Aplicación de filtros

Para poder llevar a cabo un proceso de filtrado de imagen, necesitamos un filtro, también llamado máscara (mask). El filtro es generalmente una ventana cuadrada, esa es una ventana con dos dimensiones iguales (alto y ancho). El filtro incluirá números. Esos números son llamados coeficientes, y son los que realmente determinan el efecto del filtro y cómo lucirá la imagen de salida. 

- **Filtro medio:**

    Este reemplaza básicamente cada pixel en la imagen de salida con el valor promedio de sus celdas contiguas. Esto tiene el efecto de suavizar la imagen (reduciendo la cantidad de variaciones de intensidad entre un pixel y el siguiente).
   
![alt text](ejemplo.png)
![alt text](media.png)
![alt text](media2.png)

```python
median = ndimage.median_filter(gato, size=5)
plt.imshow(median)
plt.show()
```

- **Filtro Gaussiano**

    En este filtro el elemento central representa el maximo peso que participa en la combinacion lineal de la operacion, mientras que los valores de los demas coeficientes tienen menor influencia conforme estos se alejan del centro del filtro.

    El filtro de Gauss, en comparación con el anterior, permte suavizar las regiones en donde los valores de intensidad de la luz son homogéneos sin diluir los bordes de la imagen. El valor sigma indica el nivel de desenfoque.

![alt text](gauss.png)
![alt text](gauss2.png)

```python
blurred = ndimage.gaussian_filter(gato, sigma=3) 
plt.imshow(blurred)
plt.show()
```

### Morfología binaria

Muchas veces nos va a interesar aplicar efectos solo a una parte concreta de la fotografía. Por eso son útiles las máscaras, que son matrices booleanas. De manera similar a como hacíamos en NumPy, podremos seleccionar los píxeles que nos interesen. Vamos a ver las principales herramientas que nos ayudan a agilizar el proceso.

```python
matriz = np.zeros((7, 7), dtype=np.int)
matriz[1:6, 2:5] = 1
print(matriz)
```

```python
erosion = ndimage.binary_erosion(matriz).astype(np.int)
print(erosion)
```

```python
matriz = erosion
erosion = ndimage.binary_erosion(matriz).astype(np.int)
print(erosion)
```

- **Dilation**

```python
matriz = np.zeros((5, 5)).astype(np.int)
matriz[2, 2] = 1
print(matriz)
```

```python
dilation = ndimage.binary_dilation(matriz).astype(a.dtype)
print(dilation)
```

```python
matriz = np.zeros((5, 5), dtype=np.int)
matriz[1:4, 1:4] = 1
matriz[4, 4] = 1
print(matriz)
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Elimina objetos pequeños (menores a la estructura especificada).

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Y suaviza bordes.

```python
opening = ndimage.binary_opening(matriz).astype(np.int)
print(opening)
```

- **Closing**

```python
matriz = np.zeros((5, 5), dtype=np.int)
matriz[1:4, 1:4] = 1
matriz[2,2] = 0
print(matriz)
```

Rellena huecos.

```python
closing = ndimage.binary_closing(matriz).astype(np.int)
print(closing)
```

Ahora vamos a utilizar estas herramientas en un ejemplo, 'limpiando' una imagen.

(Ejemplo de *Scipy lecture notes 2020*)

```python
# Generamos la imagen inicial aleatoria, con zonas blancas y negras distinguibles y ruido en ambas
np.random.seed(1)
n = 10
l = 256
im = np.zeros((l, l))
points = l*np.random.random((2, n**2))
im[(points[0]).astype(np.int), (points[1]).astype(np.int)] = 1
im = ndimage.gaussian_filter(im, sigma=l/(4.*n))

mask = (im > im.mean()).astype(np.float)

img = mask + 0.3*np.random.randn(*mask.shape)

binary_img = img > 0.5

# Eliminamos los puntos blancos
open_img = ndimage.binary_opening(binary_img)

# Eliminamos los agujeros negros
close_img = ndimage.binary_closing(open_img)

l = 128


plt.figure(figsize=(15, 6))

plt.subplot(141)
plt.imshow(binary_img[:l, :l], cmap=plt.cm.gray)
plt.axis('off')

plt.subplot(142)
plt.imshow(open_img[:l, :l], cmap=plt.cm.gray)
plt.axis('off')

plt.subplot(143)
plt.imshow(close_img[:l, :l], cmap=plt.cm.gray)
plt.axis('off')

plt.subplot(144)
plt.imshow(mask[:l, :l], cmap=plt.cm.gray)
plt.contour(close_img[:l, :l], [0.5], linewidths=2, colors='r')
plt.axis('off')

plt.subplots_adjust(wspace=0.02, hspace=0.3, top=1, bottom=0.1, left=0, right=1)

plt.show()
```

## Procesamiento de señales
Veamos un ejemplo para corregir la tendencia lineal de una señal. Lo primero, generamos una señal aleatoria con una tendencia evidente. Después, aplicamos la función correspondiente de SciPy y visualizamos todo el problema.

```python
from scipy import signal

t = np.linspace(0, 7, 100)
x = t + np.cos(t**2)

x_corregida = signal.detrend(x)

plt.figure(figsize=(10,10))
plt.plot(t, x, label="Señal original")
plt.plot(t, x_corregida, label="Señal corregida")
plt.plot(t,t, color = 'k', linewidth = 0.5)
plt.plot(t, np.full_like(t,0), color = 'k', linewidth = 0.5)
plt.legend(loc='best')
plt.show()
```

El siguiente ejemplo trata sobre la conversión de frecuencia de muestreo de una señal. Esto es algo así como discretizar una función, en este caso una onda, y es un proceso muy habitual cuando se trabaja con señales, por ejemplo, de audio.

```python
t = np.linspace(0, 4, 100)
x = np.sin(t**2) + np.cos(t/4)

# Vamos a extraer solo 25 puntos en vez de los 100 con los que generamos la onda original
x_resampled = signal.resample(x, 25)

plt.figure(figsize=(12, 6))
plt.plot(t, x, label='Señal original')
plt.plot(t[::4], x_resampled, 'ko', label='Señal discreta')
plt.legend(loc='best')
plt.show()
```
